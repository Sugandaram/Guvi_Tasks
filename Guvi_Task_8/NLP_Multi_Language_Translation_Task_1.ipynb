```python
def remove_duplicates(duplist):
  noduplist = []

  for element in duplist:
    if element not in noduplist:
      noduplist.append(element)
  return noduplist

print(remove_duplicates(MLT))
```

```python
from googletrans import Translator
translator = Translator()
import googletrans
```

```python
MLT_1=pd.DataFrame(MLT.iloc[0:2500,:].values, columns = ["English"])
MLT_1["Spanish"] = MLT_1['English'].apply(translator.translate, src='en', dest='es').apply(getattr, args=('text',))
MLT_1["German"] = MLT_1['English'].apply(translator.translate, src='en', dest='de').apply(getattr, args=('text',))
MLT_1["French"] = MLT_1['English'].apply(translator.translate, src='en', dest='fr').apply(getattr, args=('text',))

MLT_1.to_csv("MLT_SFG_1.csv")
```
